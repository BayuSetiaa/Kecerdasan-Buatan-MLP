```python
# Fungsi sigmoid
def sigmoid(x):
    return 1 / (1 + 2.718281828459045 ** (-x))  # 2.718281828459045 adalah nilai e

# Fungsi turunan sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Inisialisasi bobot awal
w1a, w1b = 0.24, -0.3
w2a, w2b = 0.16, 0.1
w3a, w3b = -0.2, 0.20
v1a, v2b, v3c = -0.4, 0.2, -0.3

# Bias
b = 1

# Learning rate
learning_rate = 0.5

# Data input
x1_values = [2, 3, 4, 5]
x2_values = [2, 6, 8, 1]
target_values = [0, 1, 1, 0]

# Training
for epoch in range(1000):  # Jumlah iterasi 
    total_error = 0

    for i in range(len(x1_values)):
        x1, x2 = x1_values[i], x2_values[i]
        target = target_values[i]

        # Feedforward
        z_in1 = b + x1 * w1a + x2 * w2a
        z_out1 = sigmoid(z_in1)

        z_in2 = b + x1 * w1b + x2 * w2b
        z_out2 = sigmoid(z_in2)

        y_in = b + z_out1 * v1a + z_out2 * v2b
        y_out = sigmoid(y_in)

        # Feedback (Backpropagation)
        error = 0.5 * (target - y_out) ** 2
        total_error += error

        # Hitung Error Turunan δ untuk node output y
        delta_output = (target - y_out) * sigmoid_derivative(y_out)

        # Hitung Koreksi Bobot Δ untuk setiap bobot yang tersambung ke node pada layer output y
        delta_w1a = learning_rate * delta_output * z_out1
        delta_w1b = learning_rate * delta_output * z_out2

        delta_w2a = learning_rate * delta_output * z_out1
        delta_w2b = learning_rate * delta_output * z_out2

        # Hitung Error Turunan δ untuk node hidden z
        delta_z1 = delta_output * v1a * z_out1 * (1 - z_out1)
        delta_z2 = delta_output * v2b * z_out2 * (1 - z_out2)

        # Hitung Koreksi Bobot Δ untuk setiap bobot yang tersambung ke node pada layer hidden z
        delta_v1a = learning_rate * delta_z1 * b
        delta_v2b = learning_rate * delta_z2 * b

        # Update Bobot w untuk setiap bobot yang menyambungkan layer hidden ke layer output
        w1a += delta_w1a
        w1b += delta_w1b

        w2a += delta_w2a
        w2b += delta_w2b

        # Update Bobot v untuk setiap bobot yang menyambungkan layer input ke layer hidden
        v1a += delta_v1a
        v2b += delta_v2b

    # Output error untuk setiap epoch
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Total Error = {total_error}")

# Evaluasi hasil akhir
print("\nHasil Akhir:")
for i in range(len(x1_values)):
    x1, x2 = x1_values[i], x2_values[i]
    target = target_values[i]

    z_in1 = b + x1 * w1a + x2 * w2a
    z_out1 = sigmoid(z_in1)

    z_in2 = b + x1 * w1b + x2 * w2b
    z_out2 = sigmoid(z_in2)

    y_in = b + z_out1 * v1a + z_out2 * v2b
    y_out = sigmoid(y_in)

    print(f"Data-{i + 1}: Target = {target}, Output Layer = {y_out}")
```
